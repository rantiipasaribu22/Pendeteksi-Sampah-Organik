```python
from roboflow import Roboflow
rf = Roboflow(api_key="k6GQruvZdavjLF41aGcS")
project = rf.workspace("deteksisampah").project("deteksi-sampah-organik-anorganik")
version = project.version(3)
dataset = version.download("yolov8")
```

```python
fixed_yaml = """
names:
- Sampah Organik
nc: 1
roboflow:
  license: CC BY 4.0
  project: deteksi-sampah-organik-anorganik
  url: https://universe.roboflow.com/deteksisampah/deteksi-sampah-organik-anorganik/dataset/3
  version: 3
  workspace: deteksisampah
train: train/images
val: valid/images
test: test/images
"""

with open("/content/Deteksi-Sampah-Organik-Anorganik-3/data.yaml", "w") as f:
    f.write(fixed_yaml)
```

```python
from ultralytics import YOLO

model = YOLO('yolov8s.pt')  # atau model lain

model.train(
    data="/content/Deteksi-Sampah-Organik-Anorganik-3/data.yaml",
    epochs=30,
    imgsz=640,
    batch=16
)
```
